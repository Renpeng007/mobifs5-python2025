```python
import numpy as np
import matplotlib.pyplot as plt
```

**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

```python
img=plt.imread('data/NIH3T3/im/dna-0.png')
```

**[Task 1.2 (b).]()** Show the image:

```python
plt.figure(figsize=(6, 6))
plt.imshow(img, cmap='gray')
plt.colorbar(label='Intensity')
plt.title('Original image')
plt.axis('off')
plt.show()
```

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
T = 0.1
binary = (img >= T).astype(np.uint8)

plt.figure(figsize=(6, 6))
plt.imshow(binary, cmap='gray')
plt.title(f'Binary image (T = {T})')
plt.axis('off')
plt.show()
```

```python
binary
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
gt = plt.imread('data/NIH3T3/gt/0.png')
plt.imshow(gt)
```

```python
seg = binary
intersection = (seg * gt).sum()
dice = 2 * intersection / (seg.sum() + gt.sum())

print(f"Dice coefficient: {dice:.4f}")
```
