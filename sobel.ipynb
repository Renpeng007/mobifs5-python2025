**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
def sobel_h(img):
    """Convolution with horizontal Sobel operator  gx"""
    kernel_h = (1/8) * np.array([[ 1,  0, -1],
                                 [ 2,  0, -2],
                                 [ 1,  0, -1]])
    h, w = img.shape
    gx = np.zeros_like(img, dtype=float)

    padded = np.pad(img, 1, mode='edge')

    for i in range(h):
        for j in range(w):
            region = padded[i:i+3, j:j+3]
            gx[i, j] = np.sum(region * kernel_h)
    return gx

    
def sobel_v(img):
    """Convolution with vertical Sobel operator  gy"""
    kernel_v = (1/8) * np.array([[ 1,  2,  1],
                                 [ 0,  0,  0],
                                 [-1, -2, -1]])
    h, w = img.shape
    gy = np.zeros_like(img, dtype=float)

    padded = np.pad(img, 1, mode='edge')

    for i in range(h):
        for j in range(w):
            region = padded[i:i+3, j:j+3]
            gy[i, j] = np.sum(region * kernel_v)
    return gy
```

**[Task 2.3.]()** Test your implementations from above:

```python
gx = sobel_h(img)
gy = sobel_v(img)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(gx, cmap='gray')
plt.title('gx')
plt.colorbar()
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gy, cmap='gray')
plt.title('gy')
plt.colorbar()
plt.axis('off')

plt.show()
```

**[Task 2.4.]()** Compute the image gradient magnitude:

```python
grad_mag = np.sqrt(gx**2 + gy**2)

plt.figure(figsize=(5, 5))
plt.imshow(grad_mag, cmap='gray')
plt.title('Gradient magnitude')
plt.colorbar()
plt.axis('off')
plt.show()
```
